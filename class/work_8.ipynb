# True Matrix Multiplication

$$

\begin{pmatrix}
    0 & 3 \\  
    2 & 0 \\  
\end{pmatrix} 
\times
 \begin{pmatrix}
    7 & 5 \\  
    9 & 0 \\  
\end{pmatrix}
=
\begin{pmatrix}
    0*7+3*9 & 0*5+3*0 \\  
    2*7+0*9 & 2*5+0*0 \\  
\end{pmatrix}

$$

$$
a (m \times n)  \\
b (w \times v)  \\ 
   \\   
c = a * b = (m \times v) \\ 
   \\   
if \quad and \quad only \quad if \quad n=w
$$  

```python
# the multiplication works if the column shape of first matrix or array
# is equal to the row shape of the second matrix.

axb = a @ b

print('a= \n', a)
print('b= \n', b)

array_properties(axb, 'axb')
```

```python
bxa = b @ a

print('b= \n', b)
print('a= \n', a)

array_properties(bxa, 'bxa')
```

```python
# a(2x2)
# a1(2x1)
# a column shape n = 2
# a1 row shape w = 2
# since n = w, therefore axa1 (2x1)

axa1 = a @ a1

print('a= \n', a)
print('a1= \n', a1)

array_properties(axa1, 'axa1')
```

```python
a1xa = a1 @ a

print('a1= \n', a1)
print('a= \n', a)

array_properties(a1xa, 'a1xa')
```

# Class Activity 5

System of linear equation
$$

2x_{1} + 4x_{2}  = 4 \\ 
x_{1} - 24x_{2}  = 2 \\ 

$$

To solve
$$
A.x = b  \\ 
x = A^{-1}.b
$$

```python
A = np.array([
    [2,4],
    [1,-2],
    ])

b = np.array([
    [4],
    [2],
    ])

invA = np.linalg.inv(A)

x = invA @ b

x
```

```python
# check if correct
# from equation 
2*x[0] + 4*x[1] == 4
```

```python
# alternative way
np.linalg.solve(A, b)
```

# Class Activity 6

Solve the system of linear equation below
$$

2x_{1} + 4x_{2} + x_{3}  = 4 \\ 
x_{1} - 24x_{2} + 5.5x_{3}  = 2 \\ 
2x_{1} - 4.2x_{2} + 1.5x_{3}  = -2 \\ 

$$

```python
A = np.array([
    [2, 4, 1],
    [1, -24, 5.5],
    [2, -4.2, 1.5]
])

B = np.array([
    [4],
    [2],
    [2]
])

# method 1
inverse_A = np.linalg.inv(A)

# method 2
ans1 = inverse_A @ B
ans2 = np.linalg.solve(A,B)

print('Answer for Method 1 = \n', ans1)
print('\nAnswer for Method 2 = \n', ans2)
```
