# Assignment

1. Create a 3D data of shape 10x5x3 with values equal zero.
2. Create another array woth values 255 and shape 10x5x3.
3. Join the 2 array horizontally
4. Create an array of random unsigned integer 8 bits with shape (50x50x3)
5. Convert top part to red and bottom part to green
6. Seed value is 11
7. Display or show each data


```python
import numpy as np
import matplotlib.pyplot as plt

# Q1 & Q2 Create zero amd 255 array

zero_array = np.zeros((10,5,3), dtype='uint8')
array_255 = np.full((10,5,3),255)

# Alternative ways to create array_255
# array_255 = np.where(zero_array==0,255, a)
# array_255 = np.ones((10,5,3))*255

# Q3. Join the array 
h_joined = np.hstack((zero_array, array_255))
v_joined = np.vstack((zero_array, array_255))


# Display the array and the stack difference
plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
plt.imshow(zero_array)
plt.title('Array of 0')

plt.subplot(1,4,2)
plt.imshow(array_255)
plt.title('Array of 255')

plt.subplot(1,4,3)
plt.imshow(h_joined)
plt.title('Horizontally joined')

plt.subplot(1,4,4)
plt.imshow(v_joined)
plt.title('Vertically joined')
plt.show()
```

```python
# Q4. Create an array of random unsigned integer 8 bits with shape (50x50x3)

np.random.seed(11)
b = np.random.randint(0, 256, (50, 50, 3))

plt.figure(figsize=(5,10))
plt.imshow(b)
plt.title('ORIGINAL ARRAY')
plt.show()
```

```python
# Q5. Convert top part to red (255,0,0) and bottom part to green (0,255,0)

# Identify top and bottom part
top_part = b[:26,:,:]
bottom_part = b[26:,:,:]

# Create an array of 25 rows for red and green values
red = np.full((25,50,3),[255,0,0])
green = np.full((25,50,3),[0,255,0])

# Replace the top and bottom row with red and green rows
top_part = red
bottom_part = green

# Stack red and green vertically
red_green = np.vstack((top_part, bottom_part))

# Display the result
plt.figure(figsize=(10,20))

plt.subplot(1,2,1)
plt.imshow(b)
plt.title('ORIGINAL ARRAY')

plt.subplot(1,2,2)
plt.imshow(red_green)
plt.title('RED TOP, GREEN BOTTOM')
plt.show()
```
